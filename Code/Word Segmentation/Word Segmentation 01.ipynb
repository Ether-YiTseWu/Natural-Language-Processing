```python
import re
```

```python
# read input txt and make it to str
txtInput = open('wsegInput_sample.txt',encoding="utf-8")
strInput = txtInput.read()
print(strInput)

# check it's str type
print(type(strInput))

# split
splitInput = strInput.split()
splitInput
```

```python
# read output txt and make it to str
txtOutput = open('wsegOutput_sample.txt',encoding="utf-8")
strOutput = txtOutput.read()
print(strOutput)
```

```python
# read dict txt and make it to str
txtDict = open('lexicon_sample.txt',encoding="utf-8")
strDict = txtDict.read()

# check it's str type
print(type(strDict))

# split
splitDict = strDict.split()
splitDict
```

```python
# list the probable word sets
probableSets = []
seq = 1
for word in splitInput[0]:
    for dictWords in splitDict:
        if dictWords == word or dictWords == word+splitInput[0][seq]:
            probableSets.append(dictWords)
            print(dictWords)
    if seq < (len(splitInput[0])-1):
        seq = seq + 1

# initialize the probable words sets
firstSets = ['的','的確']
secondSets = ['確','確實']
thirdSets = ['實','實行']
fourthSets = ['行','行動']
fifthSets = ['動 作','動作']
sixthSets = ['了','了解']
seventhSets = ['解 釋','解釋']

# find the single words in the begin str
strBegin = ''
seq = 1
for word in probableSets:
    if word in probableSets[seq]:
        break
    seq = seq + 1
    if len(word) == 1:
        strBegin = strBegin + word + ' '
strBegin  
```

```python
# def a function that next cell will need
def check(a,b):
    if (len(a)==2 and (a[1] == b or a[1] == b[0]) ):
        bb = ''
    else:
        bb = b  
    return bb
```

```python
answerWords1 = []

for i in firstSets:
    answer = ''
    
    for j in secondSets:
        jj = check(i,j)
        
        for k in thirdSets:
            kk = check(j,k)
                
            for x in fourthSets:
                xx = check(k,x)
                
                for y in fifthSets:
                    if (len(x)==2):
                        yy = '作'
                    else:
                        yy = y
                        
                    for z in sixthSets:      
                        
                        for s in seventhSets:
                            if (len(z)==2):
                                ss = '釋'
                            else:
                                ss = s
                              
                            answer = strBegin + i + ' ' + jj + ' ' + kk +' '+ xx + ' ' + yy + ' '  + z + ' '  + ss
                            
                            # ensure the answer is not repeat
                            if(re.sub('\s','',answer)==splitInput[0] and answer not in answerWords1):  
                                answerWords1.append(answer)
print("answerWords length :",len(answerWords1))
answerWords1
```
